```python
#loading the libaries
import pandas as pd
import numpy as np 
pd.set_option('display.max_columns', None)
import warnings
warnings.filterwarnings("ignore")
```

```python
#loading the datasets
bookings=pd.read_excel("Movie_Monk.xlsx",sheet_name='bookings')
customer=pd.read_excel("Movie_Monk.xlsx",sheet_name='customer')
due_date=pd.read_excel("Movie_Monk.xlsx",sheet_name='due date')
movies=pd.read_excel("Movie_Monk.xlsx",sheet_name='movies')
theater=pd.read_excel("Movie_Monk.xlsx",sheet_name='theater')
promocode=pd.read_excel("Movie_Monk.xlsx",sheet_name='promocode')
```

```python
#merging the bookings table along with customer table and storing it in df
df=pd.merge(bookings,customer,left_on='FK_Customer_Key',right_on='Customer_Key',how='left')
```

```python
df.head(2)
```

```python
#merging the df table along with due_date table and storing it in df2
df2=pd.merge(df,due_date,left_on='FK_Date_Key',right_on='date_key',how='left')
```

```python
#merging the df2 table along with movies table and storing it in df3
df3=pd.merge(df2,movies,left_on='FK_Movies_Key',right_on='Movie_Key',how='left')
```

```python
#merging the df3 table along with theater table and storing it in df4
df4=pd.merge(df3,theater,left_on='FK_Theater_Key',right_on='Theater_Key',how='left')
```

```python
#merging the df4 table along with promocode table and storing it in final_df
final_df=pd.merge(df4,promocode,left_on='FK_PromoCode_Key',right_on='PromoCode_Key',how='left')
```

```python
#Renaming the columns for better understanding
final_df.rename(columns = {'City_x':'City_Customer','State_x':'State_Customer','Country_x':'Country_Customer','Country_y':'Country_theater','State_y':'State_theater','City_y':'City_theater'}, inplace = True)
```

So, from the data we can uderstand that the null values of PromoCode,PromoCode_Key,FK_PromoCode_Key are null beacuse the haven't applied any promocode so we are replacing the values with 0 for int data-type and not applicable for object data-type

```python
#replacing null values with No key as there is no key applied
final_df['FK_PromoCode_Key']=final_df['FK_PromoCode_Key'].fillna('No Key')
```

```python
#replacing null values with No key as there is no key applied
final_df['PromoCode_Key']=final_df['PromoCode_Key'].fillna('No Key')
```

```python
#replacing with 0 as the there is no cancelled amount applied
final_df['CancelledAmount']=final_df['CancelledAmount'].fillna(0)
```

```python
#replacing with 0 as the there is promo code value applied
final_df['PromoCodeValue']=final_df['PromoCodeValue'].fillna(0)
```

```python
#replacing with Not availabe as there is no promocode availabe
final_df['PromoCode']=final_df['PromoCode'].fillna('Not Availabe')
```

```python
#replacing that with namur because i googled it and found the state as Namur
final_df['State_Customer']=final_df['State_Customer'].fillna('Namur')
```

```python
#re-cjecking for null values again
final_df.isnull().sum().sum()
```

```python
#adding a column for the overall profit
final_df['Overall_Profit']=((final_df['TicketsSold']-final_df['CancelledQuantity'])*final_df['Sales'])+final_df['CancelledAmount']
```
